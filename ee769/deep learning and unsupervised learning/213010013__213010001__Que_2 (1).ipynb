## here i  am importing the libraries which is used for clustering

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
from sklearn.manifold import TSNE
```

## here i am reading the csv file 

```python
data = pd.read_csv("https://www.ee.iitb.ac.in/~asethi/Dump/DataClustering.csv")
data.isnull().sum()
```

```python
plt.figure()
sns.boxplot(data['x1'])
plt.figure()
sns.boxplot(data['x2'])
plt.figure()
sns.boxplot(data['x3'])
plt.figure()
sns.boxplot(data['x4'])
```

```python
Df_transformed = np.log10(data)
```

```python
plt.figure()
sns.boxplot(Df_transformed['x1'])
plt.figure()
sns.boxplot(Df_transformed['x2'])
plt.figure()
sns.boxplot(Df_transformed['x3'])
plt.figure()
sns.boxplot(Df_transformed['x4'])
```

```python
ax = sns.distplot(x=data)
```

Here i am using elbow method find the best  no of cluster  and by seeing the graph we got to know that the no of clusters is 5.

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
KEV = KElbowVisualizer(KMeans(), k=10)
KEV.fit(Df_transformed)
KEV.show()
```

Here i am using tsne method for dimension reduction .

```python
t = TSNE(perplexity = 50)
dat = t.fit_transform(Df_transformed) 
Df_transformed['x'] = dat[:,0]
Df_transformed['y'] = dat[:,1]
sns.scatterplot(x="x",y="y",data=Df_transformed)
plt.show
```
